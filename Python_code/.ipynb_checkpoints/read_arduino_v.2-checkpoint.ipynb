```python
os.system('clear')
print ("======= List of USB devices =======")
serial_ports = glob.glob('/dev/ttyUSB*')
while(len(serial_ports)==0):
    print ("Connect the Arduino...")
    sleep(5);
    serial_ports = glob.glob('/dev/ttyUSB*')
        
for i in range(len(serial_ports)): 
    print (i, " - ", serial_ports[i])
    port = input("Choose the arduino port (e.g. 0): ")
        
ser = serial.Serial(serial_ports[int(port)], 115200)
```

```python

def main():
    global contador
    #ser.open()
    if ser.is_open:
        print("Established serial communication")
    else :
        print("Serial communication error")
    
   # ax.clear()
    print("Pressione 'Ctrl+C' para sair. \n")
    input("Pressione 'Enter' para iniciar a leitura: \n")
    
    
    
    while(True):    
        try:
            line = ser.readline().decode("utf-8")
            sleep(0.005)
            print(line)
            try:
                entry = line.split("\t")
                AcX = np.float(entry[0])
                AcY = np.float(entry[1])
                AcZ = np.float(entry[2])
                temp = np.float(entry[3])
                Gx = np.float(entry[4])
                Gy = np.float(entry[5])
                Gz = np.float(entry[6])
                
                print(AcX, AcY, AcZ, temp, Gx, Gy, Gz)
                
                read1.append(AcX) 
                read2.append(AcY)
                read3.append(AcZ)
                read4.append(Gx)
                read5.append(Gy)
                read6.append(Gz)
                read7.append(temp)
                
                
            except (ValueError):
                print("Erro de valor.")
                pass
        
        except (KeyboardInterrupt):
            now = datetime.now()
            print ("Voce pressionou Ctrl+C para interromper este programa! Seus dados foram salvos em 'Dados_%s.csv'"%(str(now)[:-7]))
            ser.close()
           # x = np.vstack((read1,read2,read3,read4,read5,read6,read7))
           # np.savetxt('Dados_%s.csv'%(str(now)[:-7]), x, delimiter=';')  
            break
```
