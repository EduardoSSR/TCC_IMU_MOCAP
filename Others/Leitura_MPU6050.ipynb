```python
i=0
L = 1000;
x = [];
Ax = [];
Ay = [];
Az = [];
temp = [];
Gx = [];
Gy = [];
Gz = [];
while(i<L):
    
    line = ser.readline().decode("utf-8")
    entry = line.split("\t");
    AcX  = entry[0];
    AcY  = entry[1];
    AcZ  = entry[2];
    Temp = entry[3];
    GyX  = entry[4];
    GyY  = entry[5];
    GyZ  = entry[6];
    
    Ax.append(AcX);
    Ay.append(AcY);
    Az.append(AcZ);
    temp.append(Temp);
    Gx.append(GyX);
    Gy.append(GyY);
    Gz.append(GyZ);
    
    #print(Ax)
   # print ("/nAcX = ", AcX, "AcY = ", AcY , "AcZ = ", AcZ)
    
   # print("/nTemp = ", Temp)
    
   # print("/nGyX = ", GyX, "GyY = ", GyY, "GyZ = ", GyZ )
    
    i+=1

print(Ax,Ay,Az,temp,Gx,Gy,Gz)  
#for i in range(L):
#    x.append(i);

#plt.plot(x,Ax)
#plt.plot(x,Ay)
#plt.plot(x,Az)
#plt.plot(x,Gx)
#plt.plot(x,Gy)
#plt.plot(x,Gz)
#plt.show()
```
